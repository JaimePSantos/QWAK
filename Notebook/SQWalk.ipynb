```python
from qwak.qwak import QWAK
from qwak.Operator import StochasticOperator

from sqwalk import SQWalker

import networkx as nx
import numpy as np
from matplotlib import pyplot as plt
import timeit
import sympy as sp
from sympy.abc import pi
from math import sqrt, ceil, pow
```

```python
n = 100
graph = nx.path_graph(n)
adj = nx.adj_matrix(graph).todense()
time_samples = 1200
initial_node = n // 2
```

```python
walker = SQWalker(np.array(adj), noise_param=0.0, sink_node=None)
```

```python
result = walker.run_walker(initial_node, time_samples)
new_state = result.final_state
```

```python
plt.plot(new_state.diag())
```

```python
n = 100
t = 12
graph = nx.cycle_graph(n)
init = [int(n / 2)]
```

```python
qwController = QWAK(graph, laplacian=False)
qwController.runWalk(t, init)
```

```python
plt.plot(qwController.getProbDistVec())
```
