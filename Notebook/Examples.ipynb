```python
import networkx as nx
import numpy as np
from matplotlib import pyplot as plt
from qwak import qwak as qw
from sympy import Matrix, gcd, div, Poly, Float, pprint
from sympy.abc import pi
from math import sqrt, ceil, pow
```

```python
n = 200
t = 30
graph = nx.cycle_graph(n)
init = [int(n / 2)]
```

```python
qwController = qw.QWAK(graph, laplacian=False)
qwController.runWalk(t, init)
```

```python
qwProbabilities = qwController.getProbDist()
qwAmplitudes = qwController.getWalk().getWalk()
```

```python
plt.plot(qwProbabilities.getProbVec())
```

```python
print(f"Mean: {qwProbabilities.mean()}\t "
          f"Moment 1: {qwProbabilities.moment(1)}\n"
          f"Moment 2: {qwProbabilities.moment(2)}\n"
          f"Stdev: {qwProbabilities.stdev()}\t"
          f"Alt Stdev: {qwProbabilities.altStdev()}\n"
          f"Survival Probability: {qwProbabilities.survivalProb(init[0]-5,init[0]+5)}\n"
          f"Inverse Part. Ratio: {qwAmplitudes.invPartRatio()}")
```
